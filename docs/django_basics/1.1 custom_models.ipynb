# Default Models Customization in Django's Auth System
Django's authentication system comes with several built-in models that handle users, groups, permissions, and content types. You can easily customize these models to fit your application's needs.

````python
# Custom User Model

from django.contrib.auth.models import AbstractUser
from django.db import models

class User(AbstractUser):
    # Add any additional fields you want for your custom user model here
    pass
    class Meta:
        db_table = 'auth_user'  
````

### Extending user model
- To customize the default User model, you can create a custom user model by extending `AbstractUser` or `AbstractBaseUser`. 
- Don't forget to add the `Meta` class with `db_table = 'auth_user'` to ensure it uses the same database table as the default user model or it will use format `yourappname_modelname`.
- Then you must add the following setting in your `settings.py` so that Django knows to use your custom user model:

```python
AUTH_USER_MODEL = 'yourapp.User' # directly adding model as a string
# OR
AUTH_USER_MODEL = UserModel # if you have imported the model at the top
```

For other models you can't just directly customize them but you can extend them or create proxy models as needed. But this will create new models rather than customizing the existing ones.
